```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[40,45,33,55,61,33,58]
min_t=[32,27,19,39,17,24,22]
avg_t=[22,12,34,27,24,18,30]
```

```python
plt.plot(days,max_t,'g+--',label="max")
plt.plot(days,min_t,'b+--',label="min")
plt.plot(days,avg_t,'r+--',label="avg")
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.legend(loc='best',shadow=True)
plt.grid()
```
